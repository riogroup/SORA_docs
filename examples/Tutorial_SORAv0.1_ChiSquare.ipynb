## ChiSquare Object Class

The ChiSquare Object Class within SORA was created to control the $\chi^2$ minimization tests and obtain the best fit values and uncertainties. Here we have some useful tasks that allows the user to get the best value and uncertainty within a sigma uncertainty value, plot and save the values. The documentation at <font color=blue>add link</font> contains the details about every step.

```python
## SORA package
from sora.extra import ChiSquare

## Usual packages
import numpy as np
import matplotlib.pyplot as pl
```

<a id="section_1"></a>
### 1. Instantiating the ChiSquare Object

The ChiSquare Object Class was not created to be instantiated by the user. Its main function is to control the fitting process and it will be resulting of the fitting tasks, such as _LightCurve.occ_lcfit()_ and _Occultation.fit_ellipse()_. This Object contains a python dictionary with the $\chi^2$ values and the associated tested parameters.

**The ChiSquare Docstring was designed to help the users. Also, each function has its Docstring containing its main purpose and the needed parameters (physical description and formats). Please, do not hesitate to use it.**

```python
ChiSquare?
```

```python
chi2, t_immersion, t_emersion = np.loadtxt('input/ascii/example_chi2.dat',unpack=True)

chisquare = ChiSquare(chi2, 496, t_immersion=t_immersion, t_emersion=t_emersion)
```

<a id="section_2"></a>
### 2. Viewing the results

First, the value that minimizes the $\chi^2$ can be obtained using the _ChiSquare.get_values()_.

```python
chisquare.get_values()
```

**All the values within a sigma limit can be found in a similar manner**

```python
chisquare.get_values(sigma=3)
```

**The best-fitted value and its uncertainties can be obtained using** _ChiSquare.get_nsigma()_

```python
chisquare.get_nsigma()
```

```python
chisquare.get_nsigma(sigma=3)
```

**Also, the number of fitted points (N) and the number of fitted parameters (P) can be obtained using** _ChiSquare.npts_ **and** _ChiSquare.nparam_

```python
print('Number of fitted points: {}'.format(chisquare.npts))
print('Number of fitted parameters: {}'.format(chisquare.nparam))
```

<a id="section_3"></a>
### 3. Saving the results and plots

The plot with the $\chi^2$ by fitted parameter can be seen using _ChiSquare.plot_chi2()_

```python
chisquare.plot_chi2()
```

**If the user choose to plot only one, it is possible to have complete control over it using** _matplotlib_ **functions**

```python
chisquare.plot_chi2('t_immersion')

pl.title('Occ. by Chariklo',fontsize=25)
pl.show()
```

**The values within the ChiSquare Object can be saved as an ASCII file using** _ChiSquare.to_file()_

```python
chisquare.to_file('output/example_chi2.dat')
```
